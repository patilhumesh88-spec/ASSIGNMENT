```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,15,25,30,50]

plt.plot(x,y)
plt.title('X vs Y')
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.grid(True)
plt.show()
```

```python
students=['John','Jane','Alice','Bob']
marks=[75,85,60,90]
colors=['pink','purple','gold','green']
plt.bar(students,marks,color=colors)
plt.title('Marks scored by students')
plt.xlabel('Students',color='black')
plt.ylabel('Marks',color='black')
plt.show()
```

```python
import matplotlib.pyplot as plt

regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# Find the index of the region with the highest revenue
highest_revenue_index = revenue.index(max(revenue))

# Create an 'explode' tuple to highlight the highest revenue slice
# A value greater than 0 will move the slice away from the center
explode = [0] * len(regions)
explode[highest_revenue_index] = 0.1 # This will 'explode' the largest slice by 0.1 radius

# Create the pie chart
plt.figure(figsize=(8, 8)) # Set a good figure size for the plot
plt.pie(revenue,
        labels=regions,
        autopct='%1.1f%%', # Format for displaying percentage values on the slices
        startangle=90,     # Start the first slice at the top (12 o'clock)
        explode=explode,   # Apply the explode effect to highlight a slice
        shadow=True)       # Add a shadow for a 3D effect

plt.title('Company Revenue Distribution by Region') # Add a title to the pie chart
plt.axis('equal')  # Ensures the pie chart is drawn as a perfect circle.

plt.show()
```

```python
import matplotlib.pyplot as plt
import random
sample_size=1000
data= [ random.randint(1,100) for _ in range(sample_size)]
plt.figure(figsize=(10,8))
plt.hist(data)
plt.title('Frequency Distribution of Random Integers (1-100)')
plt.xlabel('Integer Value')
plt.ylabel('Frequency')
plt.grid(axis='y',alpha=0.75)
plt.show()



```
