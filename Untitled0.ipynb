```python
def sigmoid(x):
  return (1/(1+np.exp(-x)))

def f_forwards(x,w1,w2): # Unindented to be a standalone function
    z1=x.dot(w1)
    a1=sigmoid(z1)
    z2=a1.dot(w2)
    a2=sigmoid(z2)
    return (a2)
```

```python
def generate_wt(x,y):
  l=[]
  for i in range(x*y):
    l.append(np.random.randn())
  return (np.array(l).reshape(x,y))
```

```python
def loss(out,Y):
  s=(np.square(out-Y))
  s=np.sum(s)/len(y)
  return(s)
```

```python
def back_prop(x,y,w1,w2,alpha):
  # Forward pass (re-calculated within backprop for gradient calculation)
  z1=x.dot(w1)
  a1=sigmoid(z1)
  # Corrected z2 calculation to use a1
  z2=a1.dot(w2)
  a2=sigmoid(z2)

  # Backward pass
  d2=(a2-y) # Error at output layer

  # Corrected d1 calculation: d1 = (d2 dot w2.T) * a1 * (1 - a1)
  # np.multiply for element-wise product; d2.dot(w2.T) has shape (1, N_hidden)
  d1=(d2.dot(w2.transpose())) * (a1*(1-a1))

  # Gradient calculation
  w1_adj=x.transpose().dot(d1)
  w2_adj=a1.transpose().dot(d2)

  # Weight update
  w1=w1 -(alpha*(w1_adj))
  w2=w2 -(alpha*(w2_adj))

  return (w1,w2)
```

```python
def train(x,y,w1,w2,alpha=0.01,epoch=10):
  acc_list=[] # Renamed `acc` to `acc_list` to avoid confusion
  epoch_losses_list=[] # Renamed `loss` to `epoch_losses_list` to avoid conflict with the `loss` function
  for j in range(epoch):
    current_epoch_batch_losses = [] # List to store losses for the current epoch's batches
    for i_idx in range(len(x)): # Loop over each sample in x
      # Corrected: f_forward -> f_forwards, x[i,w1,w2] -> x[i_idx] (w1, w2 are separate arguments)
      out=f_forwards(x[i_idx],w1,w2)
      # Corrected: los -> loss (assuming the function from cell 3MuMZz_OO9cH), Y[i] -> y[i_idx]
      current_epoch_batch_losses.append(loss(out,y[i_idx]))
      w1,w2=back_prop(x[i_idx],y[i_idx],w1,w2,alpha) # Corrected x[i] -> x[i_idx], y[i] -> y[i_idx]

    # These metrics are likely intended per epoch, not per batch inside the inner loop.
    # Moving them outside the inner loop.
    # Calculate average loss for the current epoch
    avg_epoch_loss = sum(current_epoch_batch_losses) / len(current_epoch_batch_losses) if current_epoch_batch_losses else 0
    epoch_losses_list.append(avg_epoch_loss)

    # Calculate accuracy for the current epoch (using the unconventional (1 - loss) * 100 logic)
    epoch_accuracy = (1 - avg_epoch_loss) * 100
    acc_list.append(epoch_accuracy)

    print("Epochs:",j+1,"acc:",epoch_accuracy)

  return (w1,w2,acc_list,epoch_losses_list)
```
