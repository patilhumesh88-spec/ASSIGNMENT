```python
x=data.iloc[:2:4].values
print("original data value:\n",x)
```

```python
from sklearn import preprocessing
min_max_scaler=preprocessing.MinMaxScaler(feature_range=(0,1))
x_after_min_max_scaler=min_max_scaler.fit_transform(x)
print("after min max scaler:\n",x_after_min_max_scaler)
```

```python
standardrization=preprocessing.StandardScaler()
x_after_standardization=standardrization.fit_transform(x)
print("after standard scaler:\n",x_after_standardization)
```
